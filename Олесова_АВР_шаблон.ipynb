Данные состоят из нескольких колонок:

1. Номер магазина, в котором собирались данные. Можно сделать сегментацию по номерам магазина. Принимает значения от 1 до 45
2. Дата, обозначающая конец рабочей недели, в течение которой собирались данные. Принимает значения от 1 до 45
3. Недельные продажи в американских долларах. Принимает значения от 05.02.2010 до 26.10.2012
4. Среднее значение температуры в течение недели в фаренгейтах. Принимает значения от -2.06 до 100
5. Цена на топливо в регионе магазина за литр в долларах. Принимает значения от 2.47 до 4.47
6. Индекс потребительских цен. Измеряет изменения с течением времени общего уровня цен на товары и услуги, которые конкретная группа населения приобретает, использует или платит за потребление. Принимает значения от 126 до 227
7. Уровень безработицы в регионе магазина. Принимает значения от 1 до 3.88 до 14.3

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

from etna.datasets import TSDataset
from etna.metrics import SMAPE
from etna.analysis import plot_forecast, plot_backtest
from etna.pipeline import Pipeline
```

```python
orig_df = pd.read_csv('monthly_australian_wine_sales.csv')
```

```python
orig_df['timestamp'] = pd.to_datetime(orig_df['month'])
orig_df['target'] = orig_df['sales']
orig_df.drop(columns=['month', 'sales'], inplace=True)
orig_df['segment'] = 'main'
orig_df.head()
```

```python
ts_df = TSDataset.to_dataset(orig_df)
print(type(ts_df))
ts_df.head()
```

```python
ts = TSDataset(ts_df, freq = 'MS')
```

```python
ts.plot()
```

```python
HORIZON = 8
```

```python
from etna.transforms import MeanTransform, LagTransform
```

```python
mean_8 = MeanTransform(
    in_column = 'target',
    window = 8,
    out_column='mean_08'
)
```

```python
lags = LagTransform(
    in_column='target',
    lags = list(range(8, 24)),
    out_column='lag'
)
```

```python
lags2 = LagTransform(
    in_column='target',
    lags = list(range(8, 13)),
    out_column='lag'
)
```

```python
from etna.models import CatBoostMultiSegmentModel
```

```python
model = CatBoostMultiSegmentModel(loss_function="MAE",
                                  n_estimators=600,
                                  learning_rate=0.05,
                                  depth=3,
                                  random_seed=42)
```

```python
transforms = [lags, mean_8]
```

```python
pipeline = Pipeline(model=model,
                    transforms=transforms,
                    horizon=HORIZON)
```

```python
train_ts, test_ts = ts.train_test_split(
    train_start = '1980-01-01',
    train_end = '1993-12-01',
    test_start = '1994-01-01',
    test_end = '1994-08-01'
)
```

```python
smape = SMAPE()
```

```python
from etna.metrics import MSE, R2
```

```python
mse = MSE()
r2 = R2()
```

```python
metrics_df, forecast_df, fold_info_df = pipeline.backtest(
    mode='expand',
    ts=train_ts,
    metrics=[smape, mse, r2]
)
```

```python
fold_info_df
```

```python
plot_backtest(forecast_df, train_ts, history_len=20)
```
